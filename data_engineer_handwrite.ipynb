```python
from glob import glob
```

```python
fname = glob("dataset/*/*/*.png")
fname[:10]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## X = Path -> Image -> Flatten

```python
X = np.array([plt.imread(f).flatten() for f in fname])
```

## y = Path -> split -> index ke 2

```python
y = np.array([int(f.split("\\")[2]) for f in fname])
```

```python
df = pd.DataFrame(X, columns=[f"feat_{i:04d}" for i in range(1, X.shape[1] + 1)])
df.head()
```

```python
df["label"] = y
df.head()
```

```python
df.to_csv("dataset/mnist_kw.csv", index=None)
```
